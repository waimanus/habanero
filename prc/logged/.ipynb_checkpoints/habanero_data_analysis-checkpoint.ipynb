```python
plt.figure(figsize=(12,8))
plt.plot(bme_time, bme_temp, '.', label='bme280')
plt.plot(therm_time, therm_temp, '.', label='thermocouple')
plt.xlim(260000, 290000)
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
```

The thermocouple looks like it has something weird going on, there is no reason the temperature should be getting below -75. I'm gonna exclude any points that are below that point.

```python
therm_array = []
for i,j in zip(therm_time, therm_temp):
    if j > -75:
        therm_array.append((i,j))
new_therm_time = [item[0] for item in therm_array]
new_therm_temp = [item[1] for item in therm_array]
```

```python
plt.figure(figsize=(12,8))
plt.plot(bme_time, bme_temp, '.', label='bme280')
plt.plot(new_therm_time, new_therm_temp, '.', label='thermocouple')
plt.xlim(260000, 290000)
plt.xlabel('Time (s)')
plt.ylabel('Temperature')
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
```

```python
data=[]
time_utc=[]
time_s=[]
alt=[]
with open('WiPi_gps_cropt.csv') as infile:
    for line in infile:
        formatted_line=line.strip().split(',')
        data.append(formatted_line)
        time_utc.append(formatted_line[3])
        time_s.append(formatted_line[4])
        alt.append(formatted_line[5])
        
```

```python
alt_floats=[float(item) for item in alt[2:]]
time_utc_floats=[calendar.timegm(time.strptime(item.strip(), '%Y-%m-%dT%H:%M:%S.000Z')) for item in time_utc[2:]]
```

```python
time_utc_floats
```

```python
plt.plot(time_utc_floats, alt_floats)
plt.xlabel('Time in Seconds since Epoch')
plt.ylabel('Altitude')
```
